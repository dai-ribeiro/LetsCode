Q08 - Faça um programa que peça a temperatura em graus Fahrenheit (F), transforme e mostre a temperatura em graus Celsius (°C).

°C = (5 * (F-32) / 9)

Obs: Tente também fazer um programa que faça o inverso: peça a temperatura em graus Celsius e a transforme em graus Fahrenheit.

```python
grausF = float(input('Informe a temperatura em Fahrenheit: '))
grausC = (5 * (grausF-32) / 9)

print(f'A temperatura em graus Celcius é {grausC}')
```

Q10 - Faça um programa que peça 2 números inteiros e um número real, calcule e mostre:

a) o produto do dobro do primeiro com metade do segundo.

b) a soma do triplo do primeiro com o terceiro.

c) o terceiro elevado ao cubo.


```python
num1 = int(input('Informe um número inteiro: '))
num2 = int(input('Informe outro número inteiro: '))
num3 = float(input('Informe um número real: '))

calculo1 = (num1 * 2) * (num2 / 2)
calculo2 = (num1 * 3) + num3
calculo3 = num3 ** 3

print(f'O produto do dobro do primeiro com metade do segundo é {calculo1}')
print(f'A soma do triplo do primeiro com o terceiro é {calculo2}')
print(f'O terceiro elevado ao cubo é {calculo3}')
```

Q14 - Desafio 1 - Peça para o usuário digitar uma velocidade inicial (em m/s), uma posição inicial (em m) e um instante de tempo (em s) e imprima a posição de um projétil nesse instante de tempo.

Use a fórmula matemática:

y(t) = y(0) + v(0)*t + (g*(t**2)/2)

Onde, g é a aceleração da gravidade (-10m/s²), y(t) é a posição final, y(0) é a posição inicial, v(0) é a velocidade inicial e t é o instante de tempo.

```python
velInicial = float(input('Informe a velocidade inicial em m/s: '))
posInicial = float(input('Informe a posição inicial: '))
tempo = int(input('Informe o tempo: '))
g = -10

posicao = posInicial + (velInicial * tempo) + (0.5 * g *(tempo**2))

print(f'A posição do projétil é no instante t={tempo} é {posicao}')
```

Q15 - Desafio 2 - Faça um programa que informe a data e a hora para o usuário. Para isso use a função datetime.now() do módulo datetime.

```python
import datetime

print('Data e hora atual: ', datetime.datetime.now())
```

```python
from datetime import datetime
agora = datetime.now()

ano = agora.year
mes = agora.month
dia = agora.day
hora = agora.hour
minuto = agora.minute
segundo = agora.second

print(dia, '/', mes, '/', ano, '/', ' ', hora, ":", minuto, ":", segundo, sep='')
```

```python
from datetime import datetime

agora = datetime.now()
formato = '%d/%m/%Y %H:%M:%S'
agoraFormatado = datetime.strftime(agora, formato)
print(agoraFormatado)
```
